## Gradient Descent

```python
##########################TASK 1
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

def df(x):
    # [REQUIRED] Define the derivative (slope) of your function here
    return 2 * x - 4

def f(x):
    # [REQUIRED] Define your function here
    return x**2 - 4*x + 4

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)  # [REQUIRED] Calculate the gradient (slope) at current x
    new_x = previous_x - step_multiplier * gradient  # [REQUIRED] Update x using gradient descent

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

```python
## TASK-4 : Code the same things as they were done for the first two examples
import numpy as np
import matplotlib.pyplot as plt

# Gradient Descent
def gradient_descent(x, df, learning_rate=0.1, precision=0.00001, max_iter=500):
    x_list = [x]
    slope_list = [df(x)]

    for n in range(max_iter):
        previous_x = x
        gradient = df(previous_x)
        x = previous_x - learning_rate * gradient

        step_size = abs(x - previous_x)

        x_list.append(x)
        slope_list.append(df(x))

        if step_size < precision:
            print('Loop ran this many times:', n)
            break

    return x_list, slope_list

# Define the function g(x) and its derivative dg(x)
def g(x):
    return x**2

def dg(x):
    return 2*x

# Initial values
initial_x = 3
learning_rate = 0.1
precision = 0.00001

# Perform gradient descent
x_values, slope_values = gradient_descent(initial_x, dg, learning_rate, precision)

# Plotting the function and its derivative
plt.figure(figsize=(12, 4))

# Plot the function g(x)
plt.subplot(1, 2, 1)
plt.plot(x_2, g(x_2), label='g(x)')
plt.scatter(x_values, [g(x) for x in x_values], color='red', label='Gradient Descent Path')
plt.title('Function g(x) and Gradient Descent')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.legend()

# Plot the derivative dg(x)
plt.subplot(1, 2, 2)
plt.plot(x_2, dg(x_2), label='dg(x)', color='orange')
plt.scatter(x_values, slope_values, color='red', label='Gradient Descent Slopes')
plt.title('Derivative dg(x) and Gradient Descent')
plt.xlabel('x')
plt.ylabel('dg(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

## Batch Gradient Descent with SymPy

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```
